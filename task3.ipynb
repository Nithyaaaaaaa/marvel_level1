```python
df["Age"] = df["Age"].fillna(df["Age"].median())

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["Country"] = df["Country"].fillna(df["Country"].mode()[0])

df["TotalPurchase"] = df["TotalPurchase"].fillna(df["TotalPurchase"].mean())
```

```python
df["Gender"] = df["Gender"].str.strip().str.lower()
df["Country"] = df["Country"].str.strip().str.title()
df["PreferredDevice"] = df["PreferredDevice"].str.strip().str.lower()
```

```python
df["SignupDate"] = pd.to_datetime(df["SignupDate"], errors="coerce")
df["LastLogin"] = pd.to_datetime(df["LastLogin"], errors="coerce")

df[["SignupDate", "LastLogin"]].info()
```

```python
df["Age"] = df["Age"].astype(int)
df["TotalPurchase"] = df["TotalPurchase"].astype(float)
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```
